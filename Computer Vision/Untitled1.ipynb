```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("..//image//001Bulbasaur.png")
plt.imshow(img)
```

```python
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
plt.imshow(hsv)
```

# Opening

```python
kernel = np.ones((30,25))
img_erode = cv2.erode(img,kernel)
img_open = cv2.dilate(img_erode,kernel)
plt.imshow(img_open)
plt.show()
plt.imshow(img)
```

# Closing

```python
kernel = np.ones((15,25))
img_dilate = cv2.dilate(img,kernel)
img_close = cv2.erode(img_dilate,kernel)
plt.imshow(img_close)
plt.show()
plt.imshow(img)
```

```python
layer = img.copy()
plt.imshow(layer)
plt.show()
for i in range(4):
#     layer = cv2.pyrUp(layer)
    layer = cv2.pyrDown(layer)
    plt.imshow(layer)
    plt.show()
```
