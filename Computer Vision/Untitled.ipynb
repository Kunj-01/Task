```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("..//image//001Bulbasaur.png",0)
plt.imshow(img)
```

# Simple Thresholding

```python
res, thres_b = cv2.threshold(img,150,200,cv2.THRESH_BINARY)
plt.imshow(thres_b)
```

```python
res, thres_b_i = cv2.threshold(img,150,200,cv2.THRESH_BINARY_INV)
plt.imshow(thres_b_i)
```

```python
res, thres_t = cv2.threshold(img,150,200,cv2.THRESH_TRUNC)
plt.imshow(thres_t)
```

```python
res, thres_to = cv2.threshold(img,150,200,cv2.THRESH_TOZERO)
plt.imshow(thres_to)
```

```python
res, thres_to_i = cv2.threshold(img,150,200,cv2.THRESH_TOZERO_INV)
plt.imshow(thres_to_i)
```

# Adaptive Thresholding

```python
thres_gauss = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,101,9)
plt.imshow(thres_gauss)
```

```python
thres_mean = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,71,10)
plt.imshow(thres_mean)
```

# Otsu Thresholding

```python
ret, thres_otsu = cv2.threshold(img,200,250,cv2.THRESH_TOZERO+cv2.THRESH_OTSU)
plt.imshow(thres_otsu)
```
