```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread(r"image/001Bulbasaur.png")
plt.imshow(img)
plt.show()
```

# Erosion

```python
kernel = np.ones((2,3))
kernel
```

```python
erode = cv2.erode(img,kernel)
```

```python
plt.imshow(erode)
plt.show()
plt.imshow(cv2.erode(img,kernel,iterations=3))
plt.show()
```

# Blurring

```python
gaussian = cv2.GaussianBlur(img,(101,101),1000,100,cv2.BORDER_REFLECT)
plt.imshow(gaussian)
```

```python
median = cv2.medianBlur(img,21)
plt.imshow(img)
plt.show()
plt.imshow(median)
```

```python
bilaterral = cv2.bilateralFilter(img,100,20,78)
plt.imshow(img)
plt.show()
plt.imshow(bilaterral)
```

# Scaling

```python
img_scale = cv2.resize(img, (0,0), fx=0.5,fy=0.5)
print(img.shape,img_scale.shape)
plt.imshow(img_scale)
```

# Rotation

```python
h,w = img.shape[:2]
img_rotate = cv2.getRotationMatrix2D((h/2,w/2),45,.5)
print(img_rotate)
img_rotate_final = cv2.warpAffine(img,img_rotate,(h,w))
plt.imshow(img_rotate_final)
```

# Edge Detection

```python
img_edge = cv2.Canny(img,100,100)
plt.imshow(img_edge)
```

# Shifting

```python
h,w = img.shape[:2]
img_shift = cv2.warpAffine(img,np.float32([[.5,0.2,0],[0,0.5,0]]),(h,w))
plt.imshow(img_shift)
plt.show()
plt.imshow(cv2.warpAffine(img,np.float32([[.5,0,0],[0.2,0.5,0]]),(h,w)))
```

# Dilation

```python
kernel = np.ones((10,10))
img_dilat = cv2.dilate(img,kernel)
plt.imshow(img_dilat)
```

# Thresholding

```python
ret, thres1 = cv2.threshold(img,150,240,cv2.THRESH_BINARY)
plt.imshow(thres1)
```
